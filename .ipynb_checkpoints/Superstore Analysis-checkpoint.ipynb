### RFM Analyis

```python
snapshot_date = df['Order Date'].max() + pd.Timedelta(days=1)

rfm_df = df.groupby('Customer ID').agg({
    'Order Date': lambda x: (snapshot_date - x.max()).days,
    'Order ID': 'count',
    'Sales': 'sum'})

rfm_df.rename(columns={'Order Date': 'Recency',
                    'Order ID': 'Frequency',
                    'Sales': 'Monetary'}, inplace=True)
```

```python
rfm_df['R_Score'] = pd.qcut(rfm_df['Recency'], 4, labels=[4, 3, 2, 1]) 
rfm_df['F_Score'] = pd.qcut(rfm_df['Frequency'], 4, labels=[1, 2, 3, 4]) 
rfm_df['M_Score'] = pd.qcut(rfm_df['Monetary'], 4, labels=[1, 2, 3, 4]) 
```

```python
def assign_segment(row):
    if row['R_Score'] == 4 and row['F_Score'] == 4 and row['M_Score'] == 4:
        return 'Champions'
    elif row['R_Score'] >= 3 and row['F_Score'] >= 3 and row['M_Score'] >= 3:
        return 'Loyal Customers'
    elif row['R_Score'] == 3 and row['F_Score'] == 1 and row['M_Score'] == 1:
        return 'New Customers'
    elif row['R_Score'] == 1 and row['F_Score'] == 4 and row['M_Score'] == 4:
        return 'Win Back Customers'
    elif row['R_Score'] <= 2 and row['F_Score'] <= 2 and row['M_Score'] <= 2:
        return 'At-risk Customers'
    else:
        return 'Others'
rfm_df['Customer_Segment'] = rfm_df.apply(assign_segment, axis=1)
```

### The customers which falls into specific groups of champions, loyal, at-risk & win-back are seggregated

```python
champions = list(rfm_df[rfm_df['Customer_Segment'] == 'Champions'].index)
at_risk_cust = list(rfm_df[rfm_df['Customer_Segment'] == 'Loyal Customers'].index)
loyal_cust = list(rfm_df[rfm_df['Customer_Segment'] == 'At-risk Customers'].index)
win_back_cust = list(rfm_df[rfm_df['Customer_Segment'] == 'Win Back Customers'].index)
```

### Time series analysis

#### Following graph shows the trend, seasonal and residuals according to sales

```python

df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_series = df.set_index('Order Date')['Sales'].resample('ME').sum()

decomposition = sm.tsa.seasonal_decompose(monthly_series, model='additive', period=12)

fig = decomposition.plot()
fig.set_size_inches(8, 6)
plt.show()
```

#### Applying ADF on monthly sales, it is evident from the p-value that monthly sales series is not staionary

```python
result = adfuller(monthly_series.dropna())
print("ADF Statistic:", result[0])
print("p-value:", result[1])
if result[1] <= 0.05:
    print("Series is stationary")
else:
    print("Series is not stationary")
```

#### Making the series stationary by applying 1st order difference for applying ARIMA model

```python
monthly_series_diff = monthly_series.diff().dropna()

plt.figure(figsize=(10,5))
plt.plot(monthly_series_diff, marker='o', color='red')
plt.title("Differenced Sales (to remove trend/seasonality)")
plt.show()
```

#### Following graph shows the ACF and PACF plots, from the graphs p and q are deternmined as 2 & 0 

```python
plt.figure(figsize=(10, 5))
plt.subplot(121)
plot_acf(monthly_series_diff, ax=plt.gca(), lags=20)

plt.subplot(122)
plot_pacf(monthly_series_diff, ax=plt.gca(), lags=20)
plt.title('Partial Autocorrelation Function (PACF)')
plt.tight_layout()
plt.show()
p, d, q = 2, 1, 0
```

#### Model summary 

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(monthly_series, order=(2,1,0))  
model_fit = model.fit()

print(model_fit.summary())
```

#### Forcasting and plotting using the ARIMA model

```python
# Forecast next 12 months
forecast = model_fit.get_forecast(steps=12)
forecast_index = pd.date_range(start=monthly_series.index[-1] + pd.offsets.MonthEnd(1),
                               periods=12, freq='ME')

# Confidence intervals
forecast_ci = forecast.conf_int()

plt.figure(figsize=(12,5))
plt.plot(monthly_series, label="Historical Sales")
plt.plot(forecast_index, forecast.predicted_mean, label="Forecast", color='red')
plt.fill_between(forecast_index,
                 forecast_ci.iloc[:,0],
                 forecast_ci.iloc[:,1], color='pink', alpha=0.3)
plt.title("ARIMA Forecast of Monthly Sales")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
